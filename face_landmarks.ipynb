***Detecting Location of Faces using face_recognition Library***

```
in_img1 = face_recognition.load_image_file("/content/drive/MyDrive/M.Tech Projects/Facial Expression Project/Project1/FaceLandmarkDetection/Input_Images/img1.jpg")
face_locations_in = face_recognition.face_locations(in_img1)

```

*A list of tuples of found face locations in (top, right, bottom, left) order*

```
face_locations_in
```

*`Plotting the first face detected `*

```
top, right, bottom, left = face_locations_in[0]
face_in_img1 = in_img1[top:bottom, left:right]
plt.imshow(face_in_img1)
image_save = Image.fromarray(face_in_img1)
image_save.save("/content/drive/MyDrive/M.Tech Projects/Facial Expression Project/Project1/FaceLandmarkDetection/Output_Images/cropped_face_img1.jpg")
width, height = image_save.size
print("The Cropped Image1 size is-->", width, height)
```
